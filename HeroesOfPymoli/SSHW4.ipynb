```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase = pd.read_csv(file_to_load)
test_age = purchase
top_spenders = purchase
popular_items = purchase
profitable_items = purchase
purchase.head()
```

```python
# Player Count
# Display the total number of players
totalplayercount = purchase["SN"].value_counts()
totalplayer = len(totalplayercount)
totalplayerdf = pd.DataFrame([{"Total Players":totalplayer}])
totalplayerdf

```

```python
# Purchasing Analysis (Total)
#unique items
uniqueitemscount = purchase["Item ID"].value_counts()
ui = len(uniqueitemscount)
ui
#average purchase price
averagepurchaseprice = purchase["Price"].mean()
averagepurchaseprice
#number of purchases
numberpurchases = purchase["Price"].count()
numberpurchases
#totalrevenue
totalrevenue = purchase["Price"].sum()
totalrevenue
PAT = pd.DataFrame([{"Number of Unique Items":ui,
                    "Average Price":averagepurchaseprice,
                    "Number of Purchases":numberpurchases,
                    "Total Revenue":totalrevenue}])
PAT
PAT["Average Price"] = PAT["Average Price"].map("${:.2f}".format)
PAT["Total Revenue"] = PAT["Total Revenue"].map("${:.2f}".format)
PAT
PAT = PAT[["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]
PAT
```

```python
# Gender Demographics - Percentage and Count of Male Players, Female Players, and Other / Non-Disclosed

# Percentage and Count of Male Players
male = purchase.loc[purchase["Gender"] == "Male", :]
male_num = male["SN"].value_counts()
male_num
totalmale = len(male_num)
totalmale
percentmale = round((totalmale / totalplayer)*100,2)

# Percentage and Count of Female Players

female = purchase.loc[purchase["Gender"] == "Female", :]
female_num = female["SN"].value_counts()
female_num
totalfemale = len(female_num)
totalfemale
percentfemale = round((totalfemale / totalplayer)*100,2)

# Percentage and Count of Other / Non-Disclosed
other1 = purchase.iloc[9,3]
other1
other = purchase.loc[purchase["Gender"] == "Other / Non-Disclosed", :]
other
other_num = other["SN"].value_counts()
other_num
totalother = len(other_num)
totalother
percentother = round((totalother / totalplayer)*100,2)
percentother

genderdemographics1 = pd.DataFrame([{"Gender":"Male","Total Count":totalmale,"Percentage of Players":percentmale},
                       {"Gender":"Female","Total Count":totalfemale,"Percentage of Players":percentfemale},
                       {"Gender":"Other / Non-Disclosed", "Total Count":totalother,"Percentage of Players":percentother}])
 

genderdemographics2 = genderdemographics1.set_index("Gender")
genderdemographics2
del genderdemographics2.index.name
genderdemographics2
genderdemographics2 = genderdemographics2[["Total Count","Percentage of Players"]]
genderdemographics2
```

```python
#Purchasing Analysis (Gender)
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame
genderanalysisdf = purchase.groupby("Gender")
genderanalysisdf
purcount = genderanalysisdf["Purchase ID"].count()
purcount
g_avg_pur_price = genderanalysisdf["Price"].mean()
g_avg_pur_price

total_pur_value = genderanalysisdf["Price"].sum()
total_pur_value


gender_analysis_summary_table = pd.DataFrame({"purchase count":purcount, 
                                              "Average Purchase Price":g_avg_pur_price,
                                              "Total Purchase Value":total_pur_value,
                                              "Avg Total Purchase per Person":(total_pur_value/genderdemographics2["Total Count"])})


gender_analysis_summary_table["Average Purchase Price"] = gender_analysis_summary_table["Average Purchase Price"].map("${:.2f}".format)
gender_analysis_summary_table["Total Purchase Value"] = gender_analysis_summary_table["Total Purchase Value"].map("${:.2f}".format)
gender_analysis_summary_table["Avg Total Purchase per Person"] = gender_analysis_summary_table["Avg Total Purchase per Person"].map("${:.2f}".format)
gender_analysis_summary_table
```
